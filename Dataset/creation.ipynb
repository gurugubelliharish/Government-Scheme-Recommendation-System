```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.myscheme.gov.in/search")

time.sleep(6)

schemes = []

# Scroll multiple times to load more schemes
for _ in range(25):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Collect scheme links
links = driver.find_elements(By.TAG_NAME, "a")

scheme_links = []

for link in links:
    href = link.get_attribute("href")
    if href and "/schemes/" in href:
        scheme_links.append(href)

scheme_links = list(set(scheme_links))

print("Total scheme links found:", len(scheme_links))

# Visit each scheme page
for link in scheme_links[:500]:   # limit to 500 schemes
    driver.get(link)
    time.sleep(2)

    try:
        name = driver.find_element(By.TAG_NAME, "h1").text
    except:
        name = ""

    try:
        description = driver.find_element(By.TAG_NAME, "p").text
    except:
        description = ""

    schemes.append({
        "scheme_name": name,
        "description": description,
        "link": link
    })

driver.quit()

df = pd.DataFrame(schemes)

df.drop_duplicates(inplace=True)

df.to_csv("government_schemes_dataset.csv", index=False)

print("Total schemes collected:", len(df))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.myscheme.gov.in/search")

time.sleep(10)

schemes = []

# Scroll to load more schemes
for _ in range(40):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Find scheme titles
titles = driver.find_elements(By.XPATH, "//h3")

# Find descriptions
descriptions = driver.find_elements(By.XPATH, "//p")

for i in range(len(titles)):
    try:
        name = titles[i].text
        desc = descriptions[i].text
    except:
        continue

    if name != "":
        schemes.append({
            "scheme_name": name,
            "description": desc
        })

driver.quit()

df = pd.DataFrame(schemes)

df.drop_duplicates(inplace=True)

df.to_csv("dataset.csv", index=False)

print("Total schemes collected:", len(df))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get("https://www.myscheme.gov.in/search")

time.sleep(8)

schemes = []

pages_to_scrape = 10   # change this to collect more schemes

for page in range(1, pages_to_scrape + 1):

    print("Scraping page:", page)

    time.sleep(5)

    titles = driver.find_elements(By.XPATH, "//h3")
    descriptions = driver.find_elements(By.XPATH, "//p")

    for i in range(len(titles)):
        name = titles[i].text.strip()

        if name != "":
            desc = ""
            if i < len(descriptions):
                desc = descriptions[i].text.strip()

            schemes.append({
                "scheme_name": name,
                "description": desc
            })

    # click next page
    try:
        next_button = driver.find_element(By.XPATH, "//button[contains(@aria-label,'Next')]")
        next_button.click()
    except:
        print("No more pages")
        break

    time.sleep(5)

driver.quit()

df = pd.DataFrame(schemes)

df.drop_duplicates(inplace=True)

df.to_csv("government_schemes_dataset.csv", index=False)

print("Total schemes collected:", len(df))
```
